```python
df1
```

```python
df11 = df1[df1.Class==1]
df12 = df1[df1.Class==2]
df13 = df1[df1.Class==3]
df14 = df1[df1.Class==4]
df15 = df1[df1.Class==5]
df16 = df1[df1.Class==6]
df17 = df1[df1.Class==7]

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter(df11['X1'], df11['X2'], df11['X3'], c ='blue')
ax.scatter(df12['X1'], df12['X2'], df12['X3'], c ='green')
ax.scatter(df13['X1'], df13['X2'], df13['X3'], c ='red')
ax.scatter(df14['X1'], df14['X2'], df14['X3'], c ='purple')
ax.scatter(df15['X1'], df15['X2'], df15['X3'], c ='black')
ax.scatter(df16['X1'], df16['X2'], df16['X3'], c ='pink')
ax.scatter(df17['X1'], df17['X2'], df17['X3'], c ='navy')

plt.show()
```
